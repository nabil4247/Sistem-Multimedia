```
import numpy as np
from math import pi
from scipy.io.wavfile import read as read_wav
from thinkdsp import read_wave
```

## Proses Plot Sampling

Melakukan plot proses sampling dengan `import`, `decorate` 

```
from thinkdsp import decorate
wave = read_wave ('nabil.wav')
wave.normalize()
wave.plot()
decorate(xlabel='Time (s)')
```

## Wave Menampilkan/Membuat Audio

Menampilkan Audio wave diatas `wave.make_audio()` 

```
wave.make_audio()
```

## 1. Membuat Plot Spectrum Suara

Proses menampilkan plot spectrum 2 arah (positif dan negatif) untuk melihat jumlah frequency yang dihasilkan oleh signal suara. Jika `full=Flase` maka hanya akan menampilkan frequency positif. Jika `full=True` maka hanya akan menampilkan freqency satu arah. Dengan `wave.make_spectrum()` 


```
wave.make_spectrum(full=True).plot()
decorate(xlabel='Frequency (Hz)')
```

```
wave.make_spectrum(full=False).plot()
decorate(xlabel='Frequency (Hz)')
```

## 2. Membuat Low Pass Filter

Proses menampilkan plot Low Pass Filter yang berguna untuk filtering frequency coba rubah nilai value `low_pass(1000.5)` 


```
#diambil melanjutkan dari no.1 diatas
spectrum = wave.make_spectrum(full=True)
#contoh low_pass = 5512.5
spectrum.low_pass(5512.5)
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```
filtered = spectrum.make_wave()
filtered.make_audio()
```

## 3. Membuat High Pass Filter

Proses menampilkan plot High Pass Filter  


```
#diambil melanjutkan dari no.1 diatas
spectrum = wave.make_spectrum(full=True)
#contoh high_pass = 5512.5
spectrum.high_pass(5512.5)
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```
filtered = spectrum.make_wave()
filtered.make_audio()
```

## 4. Melihat spectrum plot segments dengan warna 

Proses menampilkan plot spectrum dibagi segmentasi plot nya. Dengan filter dari `start=1`  


```
def plot_segments(original, filtered):
  start = 1
  duration = 0.001
  #menampilkan yang original dengan warna biru
  original.segment(start=start, duration=duration).plot(color='blue')
  #menampilkan yang filtered dengan warna merah
  filtered.segment(start=start, duration=duration).plot(color='red')
```

```
plot_segments(wave,filtered)
```

```
filtered = spectrum.make_wave()
filtered.make_audio()
```
